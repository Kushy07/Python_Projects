```python
# let's resample multiple columns 
# average closing, min low value, max high value, etc for entire data..basically resampling entire dataframe
# resampling entire df on weekly basis
# df.resample('W').mean() will give average value for all values

df.resample('W').agg({'Close': 'mean', 'High': 'max', 'Low': 'min', 'Volume': 'sum'})
```
